```python
import random
import csv

def generate_data(num_rows, num_cols):
    data = [[random.randint(1, 100) for _ in range(num_cols)] for _ in range(num_rows)]
    return data

f1_data = generate_data(100, 50)
f2_data = generate_data(100, 60)
f3_data = generate_data(100, 80)
f4_data = generate_data(100, 70)
f5_data = generate_data(100, 65)

with open("f1.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["col_{}".format(i) for i in range(50)])
    writer.writerows(f1_data)

with open("f2.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["col_{}".format(i) for i in range(60)])
    writer.writerows(f2_data)

with open("f3.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["col_{}".format(i) for i in range(80)])
    writer.writerows(f3_data)

with open("f4.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["col_{}".format(i) for i in range(70)])
    writer.writerows(f4_data)

with open("f5.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["col_{}".format(i) for i in range(65)])
    writer.writerows(f5_data)
```

```python
import pyspark
from functools import reduce
from pyspark.sql.functions import lit, col
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[1]").appName("SparkByExamples.com").getOrCreate()

df_list = [spark.read.csv("f{}.csv".format(i), header=True) for i in range(1, 6)]

cols = [len(df.columns) for df in df_list]
max_cols = max(cols)

df_list = [df.select(*[col(c) for c in df.columns] + [lit(None).alias("col_{}".format(i+j)) for i in range(len(df.columns), max_cols)]) for j, df in enumerate(df_list)]

df_final = reduce(lambda x, y: x.union(y), df_list)
```
